```python
# Function to generate image paths for each celebrity's subfolder
def get_image_paths(root_dir):
    # Replace with your actual logic for handling nested subfolders if needed
    celebrity_paths = [os.path.join(root_dir, subfolder) for subfolder in os.listdir(root_dir)
                      if os.path.isdir(os.path.join(root_dir, subfolder))]
    image_paths = []
    for celeb_path in celebrity_paths:
      for image in os.listdir(celeb_path):
        if image.endswith('.jpg'):
          image_path=os.path.join(celeb_path, image)
          # Get the image name without the extension
          image_name, _ = os.path.splitext(os.path.basename(image_path))
          # Remove all non-English characters from the image name
          image_name = ''.join(c for c in image_name if c.isascii())
          # Get the image directory
          image_dir = os.path.dirname(image_path)
          # Create the new image path with the updated image name
          new_image_path = os.path.join(image_dir, image_name + ".jpg")
          print("---image_path222: "+image_path)
          print("---new_image_path222: "+new_image_path)
          # Rename the image file
          os.rename(image_path, new_image_path)
          print("---image_path222: "+image_path)
          print("---new_image_path222: "+new_image_path)
          image_paths.append(new_image_path)

    return image_paths

# Load and preprocess data
```

```python
# Load images and labels
images = []
labels = []
for image_path in image_paths:
    img = cv2.imread(image_path)
    img = imutils.resize(img, width=224, height=224)  # Adjust dimensions if needed
    images.append(img)
    label = os.path.basename(os.path.dirname(image_path))
    label=label[:-1]  # Extract label from subfolder name
    labels.append(label)
```

```python
def remove_background(image):
    # Convert image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Use a pre-trained model or algorithm to segment the person
    # For example, you can use deep learning-based segmentation models like Mask R-CNN
    # Or simpler techniques like background subtraction if the background is relatively static

    # Here, let's use simple thresholding as an example
    _, thresh = cv2.threshold(gray, 240, 255, cv2.THRESH_BINARY)

    # Invert the binary image
    thresh_inv = cv2.bitwise_not(thresh)

    # Find contours
    contours, _ = cv2.findContours(thresh_inv, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Assuming the largest contour corresponds to the person
    max_contour = max(contours, key=cv2.contourArea)

    # Create a mask for the person
    mask = np.zeros_like(gray)
    cv2.drawContours(mask, [max_contour], -1, (255), thickness=cv2.FILLED)

    # Apply the mask to the original image
    person_only = cv2.bitwise_and(image, image, mask=mask)

    return person_only
```

```python
new_Images_without_background=[]
for image in images:
  new_img_without_background=remove_background(image)
  new_Images_without_background.append(new_img_without_background)
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_Images_without_background, labels, test_size=0.2, random_state=42)
```

```python
# Resize each image in X_train to the desired input shape of the model
resized_images = []
for image in X_train:
    resized_image = cv2.resize(image, (128, 128))
    resized_images.append(resized_image)

# Convert the list of resized images to a numpy array
X_train_resized = np.array(resized_images)
```

```python
# Resize each image in X_train to the desired input shape of the model
resized_images_test_list = []
for image in X_test:
    resized_image_test = cv2.resize(image, (128, 128))
    resized_images_test_list.append(resized_image_test)

# Convert the list of resized images to a numpy array
X_test_resized = np.array(resized_images_test_list)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Dropout, Flatten
import numpy as np
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.constraints import max_norm
from tensorflow.keras.callbacks import EarlyStopping
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder and fit_transform on y_train
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.fit_transform(y_test)
```

```python
unique_labels = np.unique(y_train_encoded)
print("Unique labels:", unique_labels)
```

```python
from tensorflow.keras import layers, models
num_classes = len(np.unique(y_train))
# Assuming num_classes is the number of unique classes
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(num_classes, activation='softmax')  # Adjust the number of units
])
```

```python
num_output_units = model.layers[-1].output_shape[-1]
assert num_output_units == num_classes, "Number of output units does not match the number of classes"
```
